## User Defined Functions

```python
def is_outlier(points, thresh=3.5):
    """
    Returns a boolean array with True if points are outliers and False 
    otherwise.

    Parameters:
    -----------
        points : An numobservations by numdimensions array of observations
        thresh : The modified z-score to use as a threshold. Observations with
            a modified z-score (based on the median absolute deviation) greater
            than this value will be classified as outliers.

    Returns:
    --------
        mask : A numobservations-length boolean array.

    References:
    ----------
        Boris Iglewicz and David Hoaglin (1993), "Volume 16: How to Detect and
        Handle Outliers", The ASQC Basic References in Quality Control:
        Statistical Techniques, Edward F. Mykytka, Ph.D., Editor.
        Function used from https://stackoverflow.com/a/11886564;
        Courtesy: Joe Kington (https://stackoverflow.com/users/325565/joe-kington)
    """
    points_nonan = points[~np.isnan(points)]
    if len(points.shape) == 1:
        points = points[:,None]
    median = np.median(points_nonan, axis=0)
    diff = np.sum((points - median)**2, axis=-1)
    diff = np.sqrt(diff)
    med_abs_deviation = np.median(diff[~np.isnan(diff)])

    modified_z_score = 0.6745 * diff / med_abs_deviation

    return np.array([x > thresh if ~np.isnan(x) else True for x in modified_z_score])
```

```python
def change_width_bars(ax, new_value) :
    """
    Change the width of the bars in a plot.

    Parameters:
    -----------
        ax : Plot axis object
        new_value : New value of width - in decimals (e.g. 0.2)

    Returns:
    --------
        doesn't return any value

    References:
    ----------
        Function used from https://stackoverflow.com/a/44542112;
        Courtesy: StackOverflow user jsgounot (https://stackoverflow.com/users/5016055/jsgounot)
    """
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)
```

```python
def change_fontsize_plot(ax, fontsize) :
    """
    Change the fontsize of all text in plot.

    Parameters:
    -----------
        ax : Plot axis object
        fontsize : Font size in pts

    Returns:
    --------
        doesn't return any value

    """
    for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] +
             ax.get_xticklabels() + ax.get_yticklabels()):
        item.set_fontsize(10)
```

```python
def fillnas_with_mode_median(df):
    """
    For a given dataframe, fill NA values with mode/median
    Mode is used for categoricals and median for numeric columns

    Parameters:
    -----------
        df : DataFrame object

    Returns:
    --------
        DataFrame object with NAs filled in

    """    
    if df.dtype.name in ['category', 'object']:
        return df.fillna(df.mode())
    else:
        return df.fillna(df.median())
```

```python
def remove_outliers_df(df):
    """
    For a given dataframe, remove outliers in any of the columns
    using the is_outlier function defined above.
    The is_outlier function denotes observations with a modified z-score
    (whiche in turn is based on the median absolute deviation) greater
    than 3.5 as outliers.

    Parameters:
    -----------
        df : DataFrame object

    Returns:
    --------
        DataFrame object with rows containing outliers in any
        numeric column removed.

    """
    outlier_mask = [0] * df.shape[0]
    print(len(outlier_mask))
    for col in df.columns.values:
        if df[col].dtype.name not in ['category', 'object']:
            col_outlier_mask = is_outlier(df[col])
            # print(f'{col} - {df[col].dtype.name} - {np.sum(col_outlier_mask)}')
            outlier_mask += col_outlier_mask
        # else:
        #     df[col] = df[col].cat.codes
    return df.loc[outlier_mask == 0,:]
```

```python
# extract time of day as categorical feature
def classify_time_categorical(hour_of_day):
    """
    Returns a string denoting categorical classificaiton of time of day into
    morning, afternoon, evening or night.

    Parameters:
    -----------
        hour of day : Hour of the day extracted from datetime object

    Returns:
    --------
        A string denoting the time of the day as referenced in the dict_times
        variable defined within this function.
    """
    # define dictionary with start and end times for period of day, night is anything other than this
    dict_times = {
        'morning': [6, 11],
        'afternoon': [12, 17],
        'evening': [18, 21]
    }

    for key,value in dict_times.items():
        if value[0] <= hour_of_day <= value[1]:
            return key
    return 'night'
```

## Preprocess and add date features

```python
# convert time zone to indian standard time
df['publish_time'] = df['publish_time'].dt.tz_convert('Asia/Kolkata')
```

```python
# pull in holidays feature
df_holidays_inp = pd.read_csv('data_pull/holidays_india_17_18.csv')

# convert to date format
df_holidays_inp['Date'] = pd.to_datetime(df_holidays_inp['Date'])

# drop duplicates in the hoiday list - a holiday is a holiday regardless of two or more reasons
df_holidays_inp.drop_duplicates('Date', inplace=True)

df_holidays_inp.info(verbose=True)
```

```python
# extract publish date
df['publish_date'] = pd.to_datetime(df['publish_time'].dt.date)
```

```python
# create other time-based features

# day of the week
df['publish_day'] = df['publish_time'].dt.weekday.astype('category')

# month of year
df['publish_month'] = df['publish_time'].dt.month.astype('category')
```

```python
# merge in holiday information
if 'holiday_type' in df.columns.values:
    del df['holiday_type']
df['holiday_type'] =\
    df.merge(df_holidays_inp[['Date', 'Type']].rename(columns={'Type':'holiday_type'}),
            left_on='publish_date', right_on='Date',
            how='left').drop(columns='Date').loc[:,'holiday_type'].fillna('No Holiday')

# include weekend as a holiday
df['holiday_type'] = df.apply(lambda x: 'Weekend' if (x['holiday_type'] == 'No Holiday' and x['publish_day']>= 5)\
                                                  else x['holiday_type'], axis=1).astype('category')
```

```python
# convert publish time to a categorical feature denoting time of the day
df['publish_time_cat'] = df['publish_time'].dt.hour\
                            .apply(lambda x: classify_time_categorical(x))\
                                                .astype('category')
```

### LightGBM Model Setup and GridSearch for hyperparameter tuning

```python
df.columns.values
```

```python
df['category_title'] = df['category_title'].astype('category')
```

```python
drop_cols_reg = ['video_id', 'comments_disabled', 'ratings_disabled', 'publish_time',
                 'category_id', 'dislikes', 'video_error_or_removed',
                 'title', 'tags', 'description', 'video_id', 'channel_title', 'thumbnail_link',
                 'comment_count', 'likes', 'trending_date', 'publish_date']
target_col = 'views'

# create target and feature dataframes
X = df.drop(columns=drop_cols_reg+[target_col])
y = df[target_col]

# train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) 

# model parameters
params = {'num_leaves': [10, 40, 100, 150],
          'max_depth': [3, 5, 8, 10],
          'learning_rate':[0.1, 0.05]
          }

# initialize model
model_lgreg = lgb.LGBMRegressor(n_estimators=1000, random_state=42,
                                silent=False)

# grid search parameters
gsearch = GridSearchCV(estimator=model_lgreg, param_grid=params, cv=5)
lgb_model = gsearch.fit(X=X_train, y=y_train, early_stopping_rounds=10,
                        verbose=50, eval_set=[(X_test, y_test)], eval_metric='rmse')
# # model training
# model_lgreg.fit(X_train, y_train, eval_set=[(X_test, y_test)],
#                 verbose=50, early_stopping_rounds=10, eval_metric='rmse')
```
